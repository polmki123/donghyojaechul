#### 추후에도 lag된 자료들을 쓰고 싶은 계정과목들은 여기서 미리 만든 이후에 lag를 진행한다

### #------------------------------------------------------------------------------------------------------------ #

## lag 진행

```python
lag_1quarter = df.groupby(["Symbol"]).transform(lambda x : x.shift(1))
lag_2quarter = df.groupby(["Symbol"]).transform(lambda x : x.shift(2))
lag_3quarter = df.groupby(["Symbol"]).transform(lambda x : x.shift(3))
lag_4quarter = df.groupby(["Symbol"]).transform(lambda x : x.shift(4))
```

```python
def lag_quarter_df(df, lag_quarter):
    data= df.copy()
    new_index_list = []
    for index in data.columns:
        new_index = index + " " + str(lag_quarter) + "Q lag"
        new_index_list.append(new_index)


    data.columns = new_index_list
    
    return data
```

```python
lag_1quarter = lag_quarter_df(lag_1quarter, 1)
lag_2quarter = lag_quarter_df(lag_2quarter, 2)
lag_3quarter = lag_quarter_df(lag_3quarter, 3)
lag_4quarter = lag_quarter_df(lag_4quarter, 4)
```

```python
temp = pd.concat([df, lag_1quarter, lag_2quarter, lag_3quarter, lag_4quarter], axis = 1)
```

```python
temp.head()
```

#### 필터링 조건들 여기 추가

```python
temp["시가총액"] = temp["상장주식수"] * temp["종가"]

# 포괄손익계산서
# 매출액 단위(천원)
temp["매출액 100억 이상"] = temp["매출액"] > 10000000000

#영업이익 4개 분기 모두 0 이상
temp["영업이익 4개분기 0 이상"] = (temp["영업이익"] > 0) & (temp["영업이익 1Q lag"] > 0) & (temp["영업이익 2Q lag"] > 0) & (temp["영업이익 3Q lag"] > 0)

# 영업이익률 증가
temp["영업이익률 전분기 대비 증가"] = (temp["영업이익률"] > temp["영업이익률 1Q lag"]) 
temp["영업이익률 전년도 동기 대비 증가"] = (temp["영업이익률"] > temp["영업이익률 4Q lag"]) 

# 매출액 3년 동안 증가
temp["매출액 전분기 대비 증가"] = (temp["매출액"] > temp["매출액 1Q lag"]) 
temp["매출액 전년도 동기 대비 증가"] = (temp["매출액"] > temp["매출액 4Q lag"]) 

# 시가총액 비율
# 이런 식으로 재무데이터와 주가 데이터가 혼합
# 이걸 할 수가 없네

#     temp["순이익시총비율 0.1이상"] = (temp["지배주주순이익"] / temp["시가총액"])  > 0.1
#     temp["영업이익시총비율 0.1이상"] = (temp["영업이익"] / temp["시가총액"]) > 0.1

# 영업이익 1Y lag와 비교하던지


## 재무상태표
# 자본잠식
temp["자본잠식 없음"] = temp["총자본"] > 0

# 부채비율
temp["부채비율"] = temp["총부채"] / temp["총자본"]
temp["부채비율 150% 이하"] = temp["부채비율"] < 1.5

# 자본금
temp["자본금 증가율"] = (temp["자본금"] - temp["자본금 1Q lag"]) / temp["자본금 1Q lag"]
temp["자본금 분기 변동없음"] = temp["자본금 증가율"].abs() < 0.05

# 이익잉여금
temp["이익잉여금 분기 증가"] = (temp["이익잉여금"] > temp["이익잉여금 1Q lag"]) &  (temp["이익잉여금"] > temp["이익잉여금 2Q lag"]) &  (temp["이익잉여금 1Q lag"] > temp["이익잉여금 2Q lag"])


## 현금흐름표

# 영업활동현금흐름
temp["영업활동으로인한현금흐름 판별"] = temp["영업활동으로인한현금흐름"] > 0
temp["4개분기 중 3분기 이상 영업활동으로인한현금흐름 0 초과"] = (temp["영업활동으로인한현금흐름 판별"].rolling(4).sum() > 3)

# 투자활동현금흐름
temp["투자활동으로인한현금흐름 판별"] = temp["투자활동으로인한현금흐름"] < 0
temp["4개분기 중 2분기 이상 투자활동으로인한현금흐름 0 미만"] = (temp["투자활동으로인한현금흐름 판별"].rolling(4).sum() > 2)


## 기타
# ROE가 5% 이상(수익성)
temp["ROE"] = temp["지배주주순이익"] / temp["총자본"]
temp["ROE 5퍼 이상"]  = temp["ROE"] > 0.05
```

```python
def assign_quarter(date):
    date = datetime.datetime.strptime(date, "%Y-%m-%d")
    year = date.year
    month = date.month
    day = date.day
    
    if month > 3:
        if month == 4:
            ref_year = year - 1
            ref_month = "12"
            
        elif month == 5:
            if day > 15:
                ref_year = year
                ref_month = "03"
            else:
                ref_year = year - 1
                ref_month = "12"
                
        elif (month == 6) or (month == 7):
            ref_year = year
            ref_month = "03"
            
        elif month == 8 :
            if day > 15:
                ref_month = "06"
            else:
                ref_month = "03"
            ref_year = year
            
        elif (month == 9) or (month == 10):
            ref_year = year
            ref_month = "06"
            
        elif month == 11:
            if day > 15:
                ref_month = "09"
                
            else:
                ref_month = "06"
            ref_year = year
                
        elif month == 12:
            ref_year = year
            ref_month = "09"
        
        
    else:
        ref_year = year - 1
        ref_month = "09"
        
    return str(ref_year) + "-" +ref_month
        
```
