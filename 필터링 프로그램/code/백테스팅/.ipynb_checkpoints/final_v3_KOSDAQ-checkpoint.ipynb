```python
class Portfolio:
    def __init__(self, name,  quantity, period):
        # ì£¼ì‹ì´ë¦„
        self.name = name
        # ì£¼ì‹ êµ¬ë§¤ìˆ˜ëŸ‰
        self.quantity = quantity
        # í¬íŠ¸í´ë¦¬ì˜¤ ìœ ì§€ê¸°ê°„
        self.period = period
        
    
class Back_test:
  
    def __init__(self, index, ascending, number, money = 100,fee = 0.01, size_class = "All"):
        # ë³´ìœ  í˜„ê¸ˆ, ì‹œìž‘ê¸ˆì•¡ 100
        self.money = money
        # ì¢…ëª© ê°€ì¹˜
        self.stock_value = 0
        # ì´ ê°€ì¹˜
        self.total_value = self.money + self.stock_value
        # í¬íŠ¸í´ë¦¬ì˜¤ ë‚´ ì£¼ì‹ ìˆ˜
        self.number = number
        # ê±°ëž˜ ìˆ˜ìˆ˜ë¥˜
        self.fee = fee
        # ì‚¬ìš© ì§€í‘œë¦¬ìŠ¤íŠ¸
        self.index = index
        # ì˜¤ë¦„ì°¨ìˆœ, ë‚´ë¦¼ì°¨ìˆœ, list of booleon
        self.ascending = ascending
        # ì¢…ëª©ì„ ë‹´ëŠ” í¬íŠ¸í´ë¦¬ì˜¤
        self.portfolio = []
        # í¬íŠ¸í´ë¦¬ì˜¤ ê°€ì¹˜
        self.portfolio_value = []
        # í¬íŠ¸í´ë¦¬ì˜¤ ê°œìˆ˜
        self.num_of_portfolio = 0
        # í¬íŠ¸í´ë¦¬ì˜¤ ê°œìˆ˜ë¥¼ êµ¬í•˜ê¸° ìœ„í•œ ì¹´ìš´í„°
        self.counter = 0
        self.average_portfolio_number = 0
        
        #í¬íŠ¸í´ë¦¬ì˜¤ ë²¡í„°í™”
        my_port = Portfolio(0,0,0)
        self.P = [my_port]
        
        #ëŒ€ ì¤‘ ì†Œ í˜•ì£¼
        self.size_class = size_class
        size_class = self.size_class.upper()

        if size_class == "ALL" or size_class == "A":  # ëŒ€í˜•ì£¼ ì¤‘í˜•ì£¼ ì†Œí˜•ì£¼ ë‚˜ëˆ ì„œ ì‹œë¬¼ë ˆì´ì…˜ì„ í•œë‹¤
            self.size_num = 4

        elif size_class == "BIG" or size_class == "B":
            self.size_num = 3

        elif size_class == "MEDIUM" or size_class == "M":
            self.size_num = 2

        elif size_class == "SMALL" or size_class == "S":
            self.size_num = 1

            
            
        
        
#         print("ì‚¬ìš©í•  ì§€í‘œëŠ” %s ìž…ë‹ˆë‹¤." % index)
#         print("ì‹œìž‘ ëˆì€ %s ìž…ë‹ˆë‹¤." % money)
#         print("í¬íŠ¸í´ë¦¬ì˜¤ì— ì‚°ìž…í•  ì£¼ì‹ ìˆ˜ëŠ” %s ìž…ë‹ˆë‹¤." % number)
#         print("ê±°ëž˜ìˆ˜ìˆ˜ë£ŒëŠ” %s ìž…ë‹ˆë‹¤." % fee)
    
        
        
    def order(self, portfolio, item, price, buy_or_sell): # itemì€ í´ëž˜ìŠ¤
        if buy_or_sell == "buy":
            portfolio.append(item)
            return item.quantity * price
        
        elif buy_or_sell == "sell":
            if item in portfolio:
                portfolio.remove(item)
                return item.quantity * (price) * (1 - self.fee)
            else:
                print("impossible to sell, item not in portfolio")
                return 0
          
        else:
            print("invalid value (buy_or_sell)")
            return 0
             

    
    def get_rank(self, month, index_list, ascending_list):  # ë‹¨ìˆœ ìƒìœ„ ì¶”ì¶œí•¨ìˆ˜. ê° indexë§ˆë‹¤ì˜ rankë¥¼ ë”í•˜ëŠ” ë°©ì‹
        instant_index = instant_index_list
        
        
        A = list(zip(index_list, ascending_list))

        B = df.copy()
        old = pd.DataFrame()

        for i in A:

            index_name = i[0]
            index_order = i[1]  # True or False list  # True = ë‚®ì„ìˆ˜ë¡ rank 1 False ë†’ì„ìˆ˜ë¡ rank 1
            
            

            if index_name in instant_index:  # ì¦‰ê°ì ìœ¼ë¡œ ì•Œ ìˆ˜ ìžˆëŠ” ì§€í‘œ ex) stock_price

                K = B[index_name + "_" + str(month)].rank(method = "first", ascending = index_order).sort_values()



            else:  # ê³µì‹œëŒ€ë¡œ ë‚˜ì˜¤ëŠ” ì§€í‘œ

                if month % 12 == 1 or month % 12 == 2: # 2ì›” í˜¹ì€ 3ì›”ì¸ ê²½ìš°
                    K = B[index_name + "_" + str(month-6)].rank(method = "first", ascending = index_order).sort_values()

                else:
                    K = B[index_name + "_" + str(month-4)].rank(method = "first", ascending = index_order).sort_values()


            old = pd.concat([old, K], axis = 1)


        return B.loc[old.dropna().sum(axis=1).sort_values().index]



    def get_quantile(self, month, quantile, index_list, ascending_list):
      
        instant_index = instant_index_list
        
        size_class = self.size_class
        size_num = self.size_num
        
          
        A = list(zip(index_list, ascending_list))
        B = ""
        old = pd.DataFrame()

        for i in A:
            
            rank_index_list = []
            index_name = i[0]
            index_order = i[1]  # True or False list  # True = ë‚®ì„ìˆ˜ë¡ rank 1 False ë†’ì„ìˆ˜ë¡ rank 1

            try: # preprocessed_ì‚¬ìš©í•˜ëŠ”ê²½ìš°

                interested = preprocessed[str(index_name) + "_" + str(month) + "_" + str(index_order) +"_" + size_class]
                old = pd.concat([old, interested], axis = 1)

                
            except: # ì•„ë‹Œê²½ìš°
                
                B = df.copy()
                B = B.dropna(subset = ["stock_price_" + str(month)])
                
                # ê±°ëž˜ê°€ ì •ìƒì ìœ¼ë¡œ ë˜ëŠ” ì• ë“¤ë§Œ
                B = B[B["available_" + str(month)] == 1]
                
                
                if index_name in instant_index:  # ì¸ìŠ¤í„´íŠ¸ ì¸ë±ìŠ¤ì¸ê²½ìš°
                    
                    if size_num == 4: # ëª¨ë“ ê²½ìš°
                        
                        K = B[index_name + "_" + str(month)].rank(method = "first", ascending = index_order).sort_values()
                        
                        
                    else:
                        
                        if index_name == "size_class": # size_classê°€ ë‘ë²ˆ ë“¤ì–´ê°€ë©´ ê·¸ëƒ¥ íŒ¨ìŠ¤í•¨
                            continue
                        
                        make_size =  pd.concat([B[index_name + "_" + str(month)], B["size_class_" + str(month)]], axis=1) 

                        
                        #ì¼ë‹¨ í•©ì¹˜ê³ 
                        make_size2 = make_size[make_size["size_class_" + str(month)] == size_num] # ëŒ€í˜•ì£¼ì¸ê²ƒ
                        
                        # ìˆœìœ„ë§¤ê¸°ê³ 
                        K = make_size2[index_name + "_" + str(month)].rank(method = "first", ascending = index_order).sort_values()

                        
                else: # ì¸ìŠ¤í„´íŠ¸ ì¸ë±ìŠ¤ê°€ ì•„ë‹Œ ê²½ìš° 
                    if month % 12 == 1 or month % 12 == 2: # 2ì›”ì¸ê²½ìš°
                        
                        if size_num == 4: # ëª¨ë“ ê²½ìš°
                            K = B[index_name + "_" + str(month-6)].rank(method = "first", ascending = index_order).sort_values()
                            
                            
                        else:
                            make_size =  pd.concat([B[index_name + "_" + str(month-6)], B["size_class_" + str(month)]], axis=1) #ì¼ë‹¨ í•©ì¹˜ê³ 
                            make_size2 = make_size[make_size["size_class_" + str(month)] == size_num] # ëŒ€í˜•ì£¼ì¸ê²ƒ
                            K = make_size2[index_name + "_" + str(month-6)].rank(method = "first", 
                                                                                 ascending = index_order).sort_values()

                            
                            
                            
                    else: # ê³µì‹œê¸°ì¤€ë‹¬ì¸ê²½ìš° - ì ¤ì¤‘ìš”
                        if size_num == 4: # ëª¨ë“ ê²½ìš°
                            
                            K = B[index_name + "_" + str(month-4)].rank(method = "first", ascending = index_order).sort_values()
                            
                        else:
                            make_size =  pd.concat([B[index_name + "_" + str(month-4)], B["size_class_" + str(month)]], axis=1) #ì¼ë‹¨ í•©ì¹˜ê³ 
                            make_size2 = make_size[make_size["size_class_" + str(month)] == size_num] # ëŒ€í˜•ì£¼ì¸ê²ƒ
                            K = make_size2[index_name + "_" + str(month-4)].rank(method = "first", 
                                                                                 ascending = index_order).sort_values()

                            
                            

                standard = K.quantile(quantile)
                interested = (K[K<standard])
                old = pd.concat([old, interested], axis = 1)

        return df.loc[old.dropna().index]

    
    
    def define_mode(self):
        
        mode = input("ì»¤ìŠ¤í…€ ëª¨ë“œë¥¼ ìž…ë ¥í•˜ì‹œë ¤ë©´ y, ì•„ë‹ˆë©´ ì•„ë¬´í‚¤ë‚˜ ëˆŒëŸ¬ì£¼ì„¸ìš”") # mode = y ì»¤ìŠ¤í…€ëª¨ë“œ
        sell = "X"  
        start_month = "X"
        
        if mode == "y" or mode == "Y":  # custom ë¦¬ë°¸ëŸ°ì‹± ì£¼ê¸°ì™€ ì‹œìž‘ ë‹¬
            
            
            start_month = int(input("ì‹œìž‘í•˜ê³ ìž í•˜ëŠ” ë‹¬ì„ ìž…ë ¥í•˜ì„¸ìš”. 0~107" ))
            
            sell = input("ë¦¬ë°¸ëŸ°ì‹± ì£¼ê¸°ì¸ ë‹¬(ìˆ«ìž) í˜¹ì€ ìžë™ ë¦¬ë°¸ëŸ°ì‹±(R)ì„ ìž…ë ¥í•˜ì„¸ìš”")
            
            if sell == "R" or sell == "r":
                sell = "R"
                
            else:
                sell = int(sell)  
            
            mode = "ì»¤ìŠ¤í…€ëª¨ë“œ"
            
        else:
            mode = "ê³µì‹œì£¼ê¸°ì— ë”°ë¼ ë¦¬ë°¸ëŸ°ì‹±(4,5,8,11ì›”)"
            
        return (mode, sell, start_month)  # ì»¤ìŠ¤í…€ ëª¨ë“œì¼ì‹œ (y, sell, start_month) ì•„ë‹ ì‹œ (*, "X", "X") ë¦¬í„´
    
    
    
    def rank_mode(self):
        
        while 1:
            
            mode = input("ëž­í¬ëª¨ë“œ - ë‹¨ìˆœí•©ì‚°ì‹ìœ¼ë¡œ ìˆœìœ„ë¥¼ ë§¤ê¸°ì‹œë ¤ë©´ S, ë¶„ìœ„ë¡œ ìˆœìœ„ë¥¼ ë§¤ê¸°ì‹œë ¤ë©´ Pë¥¼ ëˆŒëŸ¬ì£¼ì„¸ìš”")

            if mode =="S" or mode == "s":

                return ["sum"]

            elif mode == "P" or mode == "p":

                percent_mode = input("ìž…ë ¥ë¶„ìœ„ë¥¼ ì„ íƒí•´ì£¼ì„¸ìš” ex) 10% -> 0.1ë¡œ ëˆŒëŸ¬ì£¼ì„¸ìš”")
                percentile = float(percent_mode)

                return ["percent", percentile]

            else:
                print("ì˜¬ë°”ë¥´ì§€ ì•Šì€ ìž…ë ¥ìž…ë‹ˆë‹¤. ë‹¤ì‹œ ìž…ë ¥í•´ì£¼ì„¸ìš”.")
                
                
    def show_month_report(self, month):
        cur_value = self.money + self.stock_value
        time_year = 2013 + month//12
        time_month = (month%12) + 1

#         print( "%së…„ %sì›”" % (time_year, time_month))
#         print("ê°œì›” ìˆ˜ : ", month)
#         print("ë³´ìœ ì£¼ì‹ : ", self.P[0].name)
#         print("í˜„ìž¬ í˜„ê¸ˆ : ", self.money)
#         print("í˜„ìž¬ ì£¼ì‹ê°€ì¹˜ : ", self.stock_value)
#         print("ì´ ê°€ì¹˜ : ", cur_value)
#         print("----------------------end of month -----------------------------------------------------")

        Back_test.ending_capital = cur_value
        
        
    def get_sell_month(self, month, mode):
        if mode[0] == "ì»¤ìŠ¤í…€ëª¨ë“œ":    #  custom mode - ì‹œìž‘ë‹¬ê³¼ ë¦¬ë²Œëžœì‹± ì£¼ê¸°ë¥¼ ì„¤ì •í•´ì¤€ë‹¤
            start_month = mode[2]           # ì‹œìž‘í•˜ëŠ” ë‹¬
            sell = mode[1]

            if sell == "R": # ì»¤ìŠ¤í…€ + ìžë™ ë¦¬ë°¸ëŸ°ì‹±
                if month < start_month:
                    sell = -1
                else:
                    if month % 12 == 4 or month % 12 == 7: # 1ë¶„ê¸° ROA ì œì¼ ë†’ì€ í¬íŠ¸í´ë¦¬ì˜¤ êµ¬ì„± -> 5ì›” í˜¹ì€ 8ì›”
                        sell = 3
                    elif month % 12 == 10: # 11ì›” ê³µì‹œ
                        sell = 5
                    elif month % 12 == 3:  # 4ì›” ê³µì‹œ(ë§ˆì§€ë§‰ë¶„ê¸°)
                        sell = 1
                    # í•´ë‹¹ ë‹¬ì— êµ¬ìž…í•  ì¢…ëª©ì´ ì—†ì–´ë„ í¬íŠ¸í´ë¦¬ì˜¤ ë‚´ ì¢…ëª©ë“¤ì˜ ìž”ì—¬ ì›”ìˆ˜ ì°¨ê° ë° í¬íŠ¸í´ë¦¬ì˜¤ ê°€ì¹˜ ê³„ì‚°ì€ ì§„í–‰
                    else:
                        sell = 0

            else:
                buy_month = [i for i in range(start_month, total_month, sell)]  # start month - 104 ê¹Œì§€ì˜ sellë¡œ ë‚˜ëˆ ì§€ëŠ” ì •ìˆ˜ë“¤ì˜ ë¦¬ìŠ¤íŠ¸
                if month in buy_month:                                   # month ê°€ ì—¬ê¸° ì†í•˜ë©´
                    sell = mode[1]                                      # rebalnacing ì£¼ê¸°ë¥¼ sellë¡œ ì •í•˜ê³ 
                else:
                    sell = 0

        else:     # ê³µì‹œì£¼ê¸° ëŒ€ë¡œ
            # êµ¬ìž…í•˜ê²Œ ë  ì¢…ëª©ì˜ ë³´ìœ  ì›”ìˆ˜ë¥¼ ê²°ì •
                if month % 12 == 4 or month % 12 == 7: # 1ë¶„ê¸° ROA ì œì¼ ë†’ì€ í¬íŠ¸í´ë¦¬ì˜¤ êµ¬ì„± -> 5ì›” í˜¹ì€ 8ì›”
                    sell = 3
                elif month % 12 == 10: # 11ì›” ê³µì‹œ
                    sell = 5
                elif month % 12 == 3:  # 4ì›” ê³µì‹œ
                    sell = 1
                # í•´ë‹¹ ë‹¬ì— êµ¬ìž…í•  ì¢…ëª©ì´ ì—†ì–´ë„ í¬íŠ¸í´ë¦¬ì˜¤ ë‚´ ì¢…ëª©ë“¤ì˜ ìž”ì—¬ ì›”ìˆ˜ ì°¨ê° ë° í¬íŠ¸í´ë¦¬ì˜¤ ê°€ì¹˜ ê³„ì‚°ì€ ì§„í–‰
                else:
                    sell = 0

        return sell
    
      
    def sell_items(self, price):
        
        fill_zero = price.fillna(0) # ì—¬ê¸°ë‘ëŠ”ê²Œ ë§žë‚˜
        
        try:
            my_portfolio = self.P[0]           
            if my_portfolio.period == 0:
                sell_money = np.dot(fill_zero[my_portfolio.name], my_portfolio.quantity)
                self.money += sell_money * (1-self.fee)
                self.P[0] = Portfolio(0,0,0)
                
            else:
                self.stock_value += np.dot(fill_zero[my_portfolio.name],my_portfolio.quantity) 
            
            my_portfolio.period -= 1
            
            
        except:
            pass


    def buy_items(self, price, month, sell, ranking):
        # êµ¬ìž…í•´ì•¼ í•˜ëŠ” ë‹¬ì¸ ê²½ìš°
        if sell != 0:
            if ranking[0] == "sum":
                old = self.get_rank(month, self.index, self.ascending)
                P = old.iloc[:self.number]

            elif ranking[0] == "percent":
                P = self.get_quantile(month, ranking[1], self.index, self.ascending)

            # êµ¬ìž…í•´ì•¼ í•˜ëŠ” ì¢…ëª©, ìˆ˜ëŸ‰ ì‚°ì¶œ
            K = P.dropna(subset = ["stock_price_" + str(month)])
            items_price = K["stock_price_" + str(month)]
            
            

            if len(K) == 0 :
                return
            
            # í¬íŠ¸í´ë¦¬ì˜¤ì˜ ê°œìˆ˜ë¥¼ êµ¬í•œë‹¤  -> ì¶”í›„ í•¨ìˆ˜í™”
            self.num_of_portfolio = self.num_of_portfolio + len(K)
            self.counter = self.counter + 1
            
            self.average_portfolio_number = self.num_of_portfolio / self.counter
        
            
            self.money = self.money * ( 1- self.fee)

            items_quantity = ((self.money -1) / len(K)) / (items_price)
            items_name = K.index
                             
            my_portfolio = Portfolio(items_name, items_quantity, sell-1)
            buy_money = np.dot(price[items_name], items_quantity)  
            
            self.money -= buy_money
            self.stock_value+= buy_money

            self.P[0] = my_portfolio


    def simulation(self, start_month, end_month, mode, ranking):
        for month in range(start_month,end_month):
            # sell(ë¦¬ë²¨ëŸ°ì‹± ì£¼ê¸°)ë¥¼ êµ¬í•œë‹¤
            sell = self.get_sell_month(month, mode)
            if sell == -1:
                continue
#             print("ë¦¬ë°¸ëŸ°ì‹± ì£¼ê¸° :" , sell)                    # ì¶”í›„ë‹¤ì‹œí’€ì–´ì¤˜

            # ë³´ìœ  ì¢…ëª©ì˜ ê°€ì¹˜ ê³„ì‚°
            self.stock_value = 0
            price = df["stock_price_" + str(month)].copy()

            # íŒ”ì•„ì•¼ í•˜ëŠ” ë¬¼ê±´ë“¤ì€ íŒë‹¤
            self.sell_items(price)
            # ì‚¬ì•¼í•  ë¬¼ê±´ì´ ìžˆë‹¤ë©´ ì‚°ë‹¤
            self.buy_items(price, month, sell, ranking)
            # í¬íŠ¸í´ë¦¬ì˜¤ ê°€ì¹˜ ì €ìž¥, ì¶”í›„ ê·¸ëž˜í”„ ìž‘ì„±ì„ ìœ„í•¨
            self.portfolio_value.append(self.money + self.stock_value)
            # ì›”ë³„ ë¦¬í¬íŠ¸ë¥¼ ë³´ì—¬ì¤Œ
            self.show_month_report(month)
            my_portfolio = self.P[0]
            
        return "ìµœì¢…ê°€ì¹˜", self.money + self.stock_value, "í¬íŠ¸í´ë¦¬ì˜¤ í‰ê· ê°œìˆ˜", self.average_portfolio_number


    def get_it(self):
        cur_money = self.money
        define_mode = self.define_mode()
        rank_mode = self.rank_mode()

        # ë§¤ ë‹¬ ê±°ëž˜ë¥¼ ì§„í–‰

        mode = define_mode         # modeë¥¼ ì •í•œë‹¤
        print("ê³µì‹œì£¼ê¸° ì—¬ë¶€ : ", mode)

        ranking = rank_mode
        print("ëž­í‚¹ëª¨ë“œ : ", ranking[0])

        start_month = 5
        end_month = total_month
        self.simulation(start_month, end_month, mode, ranking)
        
    def get_it_by_quantile(self, start_month, end_month, percent):
        mode = ('ê³µì‹œì£¼ê¸°ì— ë”°ë¼ ë¦¬ë°¸ëŸ°ì‹±(4,5,8,11ì›”)', 'X', 'X')
        ranking = ['percent', percent]
        return self.simulation(start_month, end_month, mode, ranking)


    def show_graph(self):
        plt.plot(self.portfolio_value)
        plt.show()

        
```
