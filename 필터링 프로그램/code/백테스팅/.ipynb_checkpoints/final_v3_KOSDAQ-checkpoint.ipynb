```python
class Portfolio:
    def __init__(self, name,  quantity, period):
        # 주식이름
        self.name = name
        # 주식 구매수량
        self.quantity = quantity
        # 포트폴리오 유지기간
        self.period = period
        
    
class Back_test:
  
    def __init__(self, index, ascending, number, money = 100,fee = 0.01, size_class = "All"):
        # 보유 현금, 시작금액 100
        self.money = money
        # 종목 가치
        self.stock_value = 0
        # 총 가치
        self.total_value = self.money + self.stock_value
        # 포트폴리오 내 주식 수
        self.number = number
        # 거래 수수류
        self.fee = fee
        # 사용 지표리스트
        self.index = index
        # 오름차순, 내림차순, list of booleon
        self.ascending = ascending
        # 종목을 담는 포트폴리오
        self.portfolio = []
        # 포트폴리오 가치
        self.portfolio_value = []
        # 포트폴리오 개수
        self.num_of_portfolio = 0
        # 포트폴리오 개수를 구하기 위한 카운터
        self.counter = 0
        self.average_portfolio_number = 0
        
        #포트폴리오 벡터화
        my_port = Portfolio(0,0,0)
        self.P = [my_port]
        
        #대 중 소 형주
        self.size_class = size_class
        size_class = self.size_class.upper()

        if size_class == "ALL" or size_class == "A":  # 대형주 중형주 소형주 나눠서 시물레이션을 한다
            self.size_num = 4

        elif size_class == "BIG" or size_class == "B":
            self.size_num = 3

        elif size_class == "MEDIUM" or size_class == "M":
            self.size_num = 2

        elif size_class == "SMALL" or size_class == "S":
            self.size_num = 1

            
            
        
        
#         print("사용할 지표는 %s 입니다." % index)
#         print("시작 돈은 %s 입니다." % money)
#         print("포트폴리오에 산입할 주식 수는 %s 입니다." % number)
#         print("거래수수료는 %s 입니다." % fee)
    
        
        
    def order(self, portfolio, item, price, buy_or_sell): # item은 클래스
        if buy_or_sell == "buy":
            portfolio.append(item)
            return item.quantity * price
        
        elif buy_or_sell == "sell":
            if item in portfolio:
                portfolio.remove(item)
                return item.quantity * (price) * (1 - self.fee)
            else:
                print("impossible to sell, item not in portfolio")
                return 0
          
        else:
            print("invalid value (buy_or_sell)")
            return 0
             

    
    def get_rank(self, month, index_list, ascending_list):  # 단순 상위 추출함수. 각 index마다의 rank를 더하는 방식
        instant_index = instant_index_list
        
        
        A = list(zip(index_list, ascending_list))

        B = df.copy()
        old = pd.DataFrame()

        for i in A:

            index_name = i[0]
            index_order = i[1]  # True or False list  # True = 낮을수록 rank 1 False 높을수록 rank 1
            
            

            if index_name in instant_index:  # 즉각적으로 알 수 있는 지표 ex) stock_price

                K = B[index_name + "_" + str(month)].rank(method = "first", ascending = index_order).sort_values()



            else:  # 공시대로 나오는 지표

                if month % 12 == 1 or month % 12 == 2: # 2월 혹은 3월인 경우
                    K = B[index_name + "_" + str(month-6)].rank(method = "first", ascending = index_order).sort_values()

                else:
                    K = B[index_name + "_" + str(month-4)].rank(method = "first", ascending = index_order).sort_values()


            old = pd.concat([old, K], axis = 1)


        return B.loc[old.dropna().sum(axis=1).sort_values().index]



    def get_quantile(self, month, quantile, index_list, ascending_list):
      
        instant_index = instant_index_list
        
        size_class = self.size_class
        size_num = self.size_num
        
          
        A = list(zip(index_list, ascending_list))
        B = ""
        old = pd.DataFrame()

        for i in A:
            
            rank_index_list = []
            index_name = i[0]
            index_order = i[1]  # True or False list  # True = 낮을수록 rank 1 False 높을수록 rank 1

            try: # preprocessed_사용하는경우

                interested = preprocessed[str(index_name) + "_" + str(month) + "_" + str(index_order) +"_" + size_class]
                old = pd.concat([old, interested], axis = 1)

                
            except: # 아닌경우
                
                B = df.copy()
                B = B.dropna(subset = ["stock_price_" + str(month)])
                
                # 거래가 정상적으로 되는 애들만
                B = B[B["available_" + str(month)] == 1]
                
                
                if index_name in instant_index:  # 인스턴트 인덱스인경우
                    
                    if size_num == 4: # 모든경우
                        
                        K = B[index_name + "_" + str(month)].rank(method = "first", ascending = index_order).sort_values()
                        
                        
                    else:
                        
                        if index_name == "size_class": # size_class가 두번 들어가면 그냥 패스함
                            continue
                        
                        make_size =  pd.concat([B[index_name + "_" + str(month)], B["size_class_" + str(month)]], axis=1) 

                        
                        #일단 합치고
                        make_size2 = make_size[make_size["size_class_" + str(month)] == size_num] # 대형주인것
                        
                        # 순위매기고
                        K = make_size2[index_name + "_" + str(month)].rank(method = "first", ascending = index_order).sort_values()

                        
                else: # 인스턴트 인덱스가 아닌 경우 
                    if month % 12 == 1 or month % 12 == 2: # 2월인경우
                        
                        if size_num == 4: # 모든경우
                            K = B[index_name + "_" + str(month-6)].rank(method = "first", ascending = index_order).sort_values()
                            
                            
                        else:
                            make_size =  pd.concat([B[index_name + "_" + str(month-6)], B["size_class_" + str(month)]], axis=1) #일단 합치고
                            make_size2 = make_size[make_size["size_class_" + str(month)] == size_num] # 대형주인것
                            K = make_size2[index_name + "_" + str(month-6)].rank(method = "first", 
                                                                                 ascending = index_order).sort_values()

                            
                            
                            
                    else: # 공시기준달인경우 - 젤중요
                        if size_num == 4: # 모든경우
                            
                            K = B[index_name + "_" + str(month-4)].rank(method = "first", ascending = index_order).sort_values()
                            
                        else:
                            make_size =  pd.concat([B[index_name + "_" + str(month-4)], B["size_class_" + str(month)]], axis=1) #일단 합치고
                            make_size2 = make_size[make_size["size_class_" + str(month)] == size_num] # 대형주인것
                            K = make_size2[index_name + "_" + str(month-4)].rank(method = "first", 
                                                                                 ascending = index_order).sort_values()

                            
                            

                standard = K.quantile(quantile)
                interested = (K[K<standard])
                old = pd.concat([old, interested], axis = 1)

        return df.loc[old.dropna().index]

    
    
    def define_mode(self):
        
        mode = input("커스텀 모드를 입력하시려면 y, 아니면 아무키나 눌러주세요") # mode = y 커스텀모드
        sell = "X"  
        start_month = "X"
        
        if mode == "y" or mode == "Y":  # custom 리밸런싱 주기와 시작 달
            
            
            start_month = int(input("시작하고자 하는 달을 입력하세요. 0~107" ))
            
            sell = input("리밸런싱 주기인 달(숫자) 혹은 자동 리밸런싱(R)을 입력하세요")
            
            if sell == "R" or sell == "r":
                sell = "R"
                
            else:
                sell = int(sell)  
            
            mode = "커스텀모드"
            
        else:
            mode = "공시주기에 따라 리밸런싱(4,5,8,11월)"
            
        return (mode, sell, start_month)  # 커스텀 모드일시 (y, sell, start_month) 아닐 시 (*, "X", "X") 리턴
    
    
    
    def rank_mode(self):
        
        while 1:
            
            mode = input("랭크모드 - 단순합산식으로 순위를 매기시려면 S, 분위로 순위를 매기시려면 P를 눌러주세요")

            if mode =="S" or mode == "s":

                return ["sum"]

            elif mode == "P" or mode == "p":

                percent_mode = input("입력분위를 선택해주세요 ex) 10% -> 0.1로 눌러주세요")
                percentile = float(percent_mode)

                return ["percent", percentile]

            else:
                print("올바르지 않은 입력입니다. 다시 입력해주세요.")
                
                
    def show_month_report(self, month):
        cur_value = self.money + self.stock_value
        time_year = 2013 + month//12
        time_month = (month%12) + 1

#         print( "%s년 %s월" % (time_year, time_month))
#         print("개월 수 : ", month)
#         print("보유주식 : ", self.P[0].name)
#         print("현재 현금 : ", self.money)
#         print("현재 주식가치 : ", self.stock_value)
#         print("총 가치 : ", cur_value)
#         print("----------------------end of month -----------------------------------------------------")

        Back_test.ending_capital = cur_value
        
        
    def get_sell_month(self, month, mode):
        if mode[0] == "커스텀모드":    #  custom mode - 시작달과 리벌랜싱 주기를 설정해준다
            start_month = mode[2]           # 시작하는 달
            sell = mode[1]

            if sell == "R": # 커스텀 + 자동 리밸런싱
                if month < start_month:
                    sell = -1
                else:
                    if month % 12 == 4 or month % 12 == 7: # 1분기 ROA 제일 높은 포트폴리오 구성 -> 5월 혹은 8월
                        sell = 3
                    elif month % 12 == 10: # 11월 공시
                        sell = 5
                    elif month % 12 == 3:  # 4월 공시(마지막분기)
                        sell = 1
                    # 해당 달에 구입할 종목이 없어도 포트폴리오 내 종목들의 잔여 월수 차감 및 포트폴리오 가치 계산은 진행
                    else:
                        sell = 0

            else:
                buy_month = [i for i in range(start_month, total_month, sell)]  # start month - 104 까지의 sell로 나눠지는 정수들의 리스트
                if month in buy_month:                                   # month 가 여기 속하면
                    sell = mode[1]                                      # rebalnacing 주기를 sell로 정하고
                else:
                    sell = 0

        else:     # 공시주기 대로
            # 구입하게 될 종목의 보유 월수를 결정
                if month % 12 == 4 or month % 12 == 7: # 1분기 ROA 제일 높은 포트폴리오 구성 -> 5월 혹은 8월
                    sell = 3
                elif month % 12 == 10: # 11월 공시
                    sell = 5
                elif month % 12 == 3:  # 4월 공시
                    sell = 1
                # 해당 달에 구입할 종목이 없어도 포트폴리오 내 종목들의 잔여 월수 차감 및 포트폴리오 가치 계산은 진행
                else:
                    sell = 0

        return sell
    
      
    def sell_items(self, price):
        
        fill_zero = price.fillna(0) # 여기두는게 맞나
        
        try:
            my_portfolio = self.P[0]           
            if my_portfolio.period == 0:
                sell_money = np.dot(fill_zero[my_portfolio.name], my_portfolio.quantity)
                self.money += sell_money * (1-self.fee)
                self.P[0] = Portfolio(0,0,0)
                
            else:
                self.stock_value += np.dot(fill_zero[my_portfolio.name],my_portfolio.quantity) 
            
            my_portfolio.period -= 1
            
            
        except:
            pass


    def buy_items(self, price, month, sell, ranking):
        # 구입해야 하는 달인 경우
        if sell != 0:
            if ranking[0] == "sum":
                old = self.get_rank(month, self.index, self.ascending)
                P = old.iloc[:self.number]

            elif ranking[0] == "percent":
                P = self.get_quantile(month, ranking[1], self.index, self.ascending)

            # 구입해야 하는 종목, 수량 산출
            K = P.dropna(subset = ["stock_price_" + str(month)])
            items_price = K["stock_price_" + str(month)]
            
            

            if len(K) == 0 :
                return
            
            # 포트폴리오의 개수를 구한다  -> 추후 함수화
            self.num_of_portfolio = self.num_of_portfolio + len(K)
            self.counter = self.counter + 1
            
            self.average_portfolio_number = self.num_of_portfolio / self.counter
        
            
            self.money = self.money * ( 1- self.fee)

            items_quantity = ((self.money -1) / len(K)) / (items_price)
            items_name = K.index
                             
            my_portfolio = Portfolio(items_name, items_quantity, sell-1)
            buy_money = np.dot(price[items_name], items_quantity)  
            
            self.money -= buy_money
            self.stock_value+= buy_money

            self.P[0] = my_portfolio


    def simulation(self, start_month, end_month, mode, ranking):
        for month in range(start_month,end_month):
            # sell(리벨런싱 주기)를 구한다
            sell = self.get_sell_month(month, mode)
            if sell == -1:
                continue
#             print("리밸런싱 주기 :" , sell)                    # 추후다시풀어줘

            # 보유 종목의 가치 계산
            self.stock_value = 0
            price = df["stock_price_" + str(month)].copy()

            # 팔아야 하는 물건들은 판다
            self.sell_items(price)
            # 사야할 물건이 있다면 산다
            self.buy_items(price, month, sell, ranking)
            # 포트폴리오 가치 저장, 추후 그래프 작성을 위함
            self.portfolio_value.append(self.money + self.stock_value)
            # 월별 리포트를 보여줌
            self.show_month_report(month)
            my_portfolio = self.P[0]
            
        return "최종가치", self.money + self.stock_value, "포트폴리오 평균개수", self.average_portfolio_number


    def get_it(self):
        cur_money = self.money
        define_mode = self.define_mode()
        rank_mode = self.rank_mode()

        # 매 달 거래를 진행

        mode = define_mode         # mode를 정한다
        print("공시주기 여부 : ", mode)

        ranking = rank_mode
        print("랭킹모드 : ", ranking[0])

        start_month = 5
        end_month = total_month
        self.simulation(start_month, end_month, mode, ranking)
        
    def get_it_by_quantile(self, start_month, end_month, percent):
        mode = ('공시주기에 따라 리밸런싱(4,5,8,11월)', 'X', 'X')
        ranking = ['percent', percent]
        return self.simulation(start_month, end_month, mode, ranking)


    def show_graph(self):
        plt.plot(self.portfolio_value)
        plt.show()

        
```

```python
import itertools
import time


def get_ascending(len):
    ascending_list = []
    for element in itertools.product([True, False], repeat=len):
        ascending_list.append(list(element))
        
    return ascending_list
        
        
def get_index(simulation_index, len):
    got_index_list = []
    comb = itertools.combinations(simulation_index, len)
    for index_comb in comb:
        got_index_list.append(list(index_comb))
        
    return got_index_list


def save_file(data, length, percentile, size_class = "ALL"):
    
    if size_class == "ALL":
        
        f = open("results\\{0}factor\\kosdaq\\{1}percentile.txt".format(length, percentile) , "a")
        f.write(str(data))
        f.write("\n")
        f.close()
    

    else:
        f = open("results\\{0}factor\\kosdaq\\{1}\\{2}percentile.txt".format(length,size_class, percentile) , "a")
        f.write(str(data))
        f.write("\n")
        f.close()
        
    

def file_continue(factor_num, percentile, size_class = "All"):
    
    if size_class == "All":
        
    
        try :
            f = open("results\\{0}factor\\kosdaq\\{1}percentile.txt".format(factor_num, percentile) , "r")

        except :
            print("파일이 존재하지 않습니다. 새 파일 생성후 시작")

            counter = 0

            return counter

        A = f.readlines()[-1]
        B = A.split(",")[-1]
        C = B.split("]")[0]
        counter = int(C)

        print(counter)
        
    else:
        try :
            f = open("results\\{0}factor\\kosdaq\\{1}\\{2}percentile.txt".format(factor_num, size_class, percentile) , "r")

        except :
            print("파일이 존재하지 않습니다. 새 파일 생성후 시작")

            counter = 0

            return counter

        A = f.readlines()[-1]
        B = A.split(",")[-1]
        C = B.split("]")[0]
        counter = int(C)

        print(counter)
        

        return counter


def get_4factor_simulation(percentile):
    quantile_preprocessing(5, total_month, percentile/10)
    counter = 0
    last_line = file_continue(4,percentile/10)

    all_num = len(get_index(index_list,4)) * len(get_ascending(4)) * (6-1)
    
    
    f = open("start_end\\start_end.txt" , "r")
    line = f.readline()
    f.close()
    start_end = line.split(",")
    start = int(start_end[0])
    end = int(start_end[1])
    
    
    print("factor안에 start", start)


    for index in get_index(index_list, 4):
        for order in get_ascending(4):

            if counter < last_line:

                counter = counter + 1

                pass

            else:
                counter = counter + 1
                A = [Back_test(index, order, 20).get_it_by_quantile(start, end, percentile/10), index, order, counter]
                save_file(A,4, percentile/10)


            if counter % 10 == 0:
                
                print("진행중인 percentile", percentile)
                print("사용지표 : ", index)
                print("현재 진행상황 : ", round((counter / all_num),3), "%")



def get_3factor_simulation(percentile, size_class = "SMALL"):
    quantile_preprocessing(5, total_month, percentile/10, size_class)
    counter = 0
    last_line = file_continue(3,percentile/10, size_class=size_class)

    all_num = len(get_index(index_list,3)) * len(get_ascending(3)) * (6-1)
    
    
    f = open("start_end\\start_end.txt" , "r")
    line = f.readline()
    f.close()
    start_end = line.split(",")
    start = int(start_end[0])
    end = int(start_end[1])
    
    
    print("factor안에 start", start)


    for index in get_index(index_list, 3):
        for order in get_ascending(3):

            if counter < last_line:

                counter = counter + 1

                pass

            else:
                counter = counter + 1
                A = [Back_test(index, order, 20, size_class = size_class)
                     .get_it_by_quantile(start, end, percentile/10), index, order, counter]
                save_file(A,3, percentile/10, size_class=size_class)


            if counter % 10 == 0:
                
                print("진행중인 percentile", percentile)
                print("사용지표 : ", index)
                print("현재 진행상황 : ", round((counter / all_num),3), "%")
                
                
                
def get_2factor_simulation(percentile, size_class="SMALL"):

        
    quantile_preprocessing(5, total_month, percentile/10, size_class)
    counter = 0
    last_line = file_continue(2,percentile/10, size_class=size_class)

    all_num = len(get_index(index_list,2)) * len(get_ascending(2)) * (6-1)
    
    f = open("start_end\\start_end.txt" , "r")
    line = f.readline()
    f.close()
    start_end = line.split(",")
    start = int(start_end[0])
    end = int(start_end[1])
    


    for index in get_index(index_list, 2):
        for order in get_ascending(2):

            if counter < last_line:

                counter = counter + 1

                pass

            else:
                counter = counter + 1
                A = [Back_test(index, order, 20, size_class=size_class)
                     .get_it_by_quantile(start, end, percentile/10), index, order, counter]
                save_file(A,2, percentile/10, size_class=size_class)


            if counter % 10 == 0:
                
                print("진행중인 percentile", percentile)
                print("사용지표 : ", index)
                print("현재 진행상황 : ", round((counter / all_num),3), "%")
                
                
def quantile_preprocessing(start_month, end_month, percent, size_class = "All"):
    instant_index = instant_index_list
    
    size_class = size_class.upper()
    
    if size_class == "ALL" or size_class == "A":  # 대형주 중형주 소형주 나눠서 시물레이션을 한다
        size_num = 4
    
    elif size_class == "BIG" or size_class == "B":
        size_num = 3
        
    elif size_class == "MEDIUM" or size_class == "M":
        size_num = 2
        
    elif size_class == "SMALL" or size_class == "S":
        size_num = 1
        

    for i, index in enumerate(index_list):
        print( str(i) +  ' / ' + str(len(index_list)) )

        for month in range(start_month, end_month):
            
            
            
#             이게 여기 맞나??? ## 이거부터!
            B = df.copy()
            B = B.dropna(subset = ["stock_price_" + str(month)])

            # 거래가 정상적으로 되는 애들만
            B = B[B["available_" + str(month)] == 1]


            for boolean in [True, False]:
                index_name = index
                index_order = boolean
                
                

                if index_name in instant_index:  # 인스턴트 인덱스인경우
                    
                    if size_num == 4: # 모든경우
                        
                        K = B[index_name + "_" + str(month)].rank(method = "first", ascending = index_order).sort_values()
                        
                        
                    else:
                        
                        if index_name == "size_class": # size_class가 두번 들어가면 그냥 패스함
                            continue
                        
                        make_size =  pd.concat([B[index_name + "_" + str(month)], B["size_class_" + str(month)]], axis=1) 
                        #일단 합치고
                        make_size2 = make_size[make_size["size_class_" + str(month)] == size_num] # 대형주인것
                        K = make_size2[index_name + "_" + str(month)].rank(method = "first", ascending = index_order).sort_values()
                        

             

                else: # 인스턴트 인덱스가 아닌 경우 
                    if month % 12 == 1 or month % 12 == 2: # 2월인경우
                        
                        if size_num == 4:
                            K = B[index_name + "_" + str(month-6)].rank(method = "first", ascending = index_order).sort_values()
                            
                            
                        else:
                            make_size =  pd.concat([B[index_name + "_" + str(month-6)], B["size_class_" + str(month)]], axis=1) #일단 합치고
                            make_size2 = make_size[make_size["size_class_" + str(month)] == size_num] # 대형주인것
                            K = make_size2[index_name + "_" + str(month-6)].rank(method = "first", 
                                                                                 ascending = index_order).sort_values()

                            
                            
                            
                    else: # 공시기준달인경우 - 젤중요
                        if size_num == 4:
                            
                            K = B[index_name + "_" + str(month-4)].rank(method = "first", ascending = index_order).sort_values()
                            
                        else:
                            make_size =  pd.concat([B[index_name + "_" + str(month-4)], B["size_class_" + str(month)]], axis=1) #일단 합치고
                            make_size2 = make_size[make_size["size_class_" + str(month)] == size_num] # 대형주인것
                            K = make_size2[index_name + "_" + str(month-4)].rank(method = "first", 
                                                                                 ascending = index_order).sort_values()

                            
                            

                K = K.dropna()

                preprocessed[str(index_name) + "_" + str(month) + "_" + str(boolean) + "_" + size_class] = K[ K < 
                                                                                                             K.quantile(percent)]  


    
    
   
```
